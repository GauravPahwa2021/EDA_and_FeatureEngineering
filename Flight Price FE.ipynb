## Feature Engineering 

# Handling Numerical Features And Missing Values

```python
## split date,month,year from Date_of_Journey - By str
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
## split date,month,year from Date_of_Journey - By lambda function
df['Date'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
df['Month'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
df['Year'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.info()
```

```python
## Drop feature Date_of_Journey
df.drop(['Date_of_Journey'],axis = 1,inplace = True)
df.head()
```

```python
## split Arrival_Time - By str
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
df.head(1)
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute'] = df['Arrival_Time'].str.split(':').str[1]
df.head(1)
```

```python
## Drop feature Arrival_Time
df.drop(['Arrival_Time'],axis = 1,inplace = True)
df.head()
```

```python
## split Dep_Time - By str
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_minute'] = df['Dep_Time'].str.split(':').str[1]
df.head(1)
```

```python
## Drop feature Dep_Time
df.drop(['Dep_Time'],axis = 1,inplace = True)
df.head()
```

```python
df.info()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_minute'] = df['Arrival_minute'].astype(int)
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_minute'] = df['Dep_minute'].astype(int)
df.info()
```

```python
df.isnull().sum()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
df.head()
```

```python
df.info()
```

```python
## Drop non valuable feature Route
df.drop(['Route'],axis = 1,inplace = True)
df.head(2)
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops'] = df['Total_Stops'].fillna(1)
```

```python
df['Additional_Info'].unique()
```

```python
df['duration_minute'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.head(1)
```
